# Data Wrangling/reading

```python
adt = pd.read_csv(r"C:\Users\CTTC\Downloads\Datasets\adult.data",header=None,delimiter=' *, *',engine='python')
adt
#'header' makes the first row to come to the dataset adn not get transfered to the column heading
#'delimiter' removes the blank spaces from front and back of each input data
#'engine' defines the kernel engine as 'python'
```

# Data Cleaning

```python
adt.isnull().sum()
#checks if any NULL values are present
```

```python
d1=adt.describe(include='all')
d1         #d1 stores the description of all the columns of the dataset
```

```python
adt.dtypes
#data type of each dataset column
```

```python
for i in adt.columns:
    print(i,':',sum(adt[i]=='?'))

#checks if any '?' is present in any of the columns
```

```python
for i in adt.columns:
    print(i,':',sum(adt[i]=='?'))
    
#checks if any '?' is present in any of the columns
```

```python
#heatmap-> to view the linear correlation of different attributes with each other 

cor=adt.corr()     #stores the correlation between different attributes
sns.heatmap(cor,annot=True,cmap='YlGnBu')
plt.title('Heat Map')
plt.show()

#it can be seen that as our output column is in 'object' type, thus not present in heatmap
#heatmaps are generally not preferred for 'object' type of outputs
#the correlation displayed on the heatmap is linear, thus an attribute with low correlation can also be important
```

# Label Encoding

```python
from sklearn import preprocessing
le1 = preprocessing.LabelEncoder()           #label encoder variable for 'work_class'
adt.work_class = le1.fit_transform(adt.work_class)

le2 = preprocessing.LabelEncoder()           #label encoder variable for 'education'
adt.education = le2.fit_transform(adt.education)

le3 = preprocessing.LabelEncoder()           #label encoder variable for 'maritial_status'
adt.maritial_status = le3.fit_transform(adt.maritial_status)

le4 = preprocessing.LabelEncoder()           #label encoder variable for 'occupation'
adt.occupation = le4.fit_transform(adt.occupation)

le5 = preprocessing.LabelEncoder()           #label encoder variable for 'relationship'
adt.relationship = le5.fit_transform(adt.relationship)

le6 = preprocessing.LabelEncoder()           #label encoder variable for 'race'
adt.race = le6.fit_transform(adt.race)

le7 = preprocessing.LabelEncoder()           #label encoder variable for 'sex'
adt.sex = le7.fit_transform(adt.sex)

le8 = preprocessing.LabelEncoder()           #label encoder variable for 'native_country'
adt.native_country = le8.fit_transform(adt.native_country)

le8 = preprocessing.LabelEncoder()           #label encoder variable for 'Salary'
adt.Salary = le8.fit_transform(adt.Salary)

#it is always a good practice to use separate encoder variables, as then the data stored can easily be inversed
```

# Ip/Op creation

```python
ip=adt.drop('Salary',axis=1)
op=adt.Salary
```

# ML Model

```python
maxia=0       #stores the random_state at which the accuracy is maximum
maxip=0       #stores the random_state at which the precision is maximum
maxir=0       #stores the random_state at which the recall is maximum
maxacc=0      #stores the value of maximum accuracy
maxpre=0      #stores the value of maximum precision
maxrec=0      #stores the value of maximum recall
for i in range(0,501):                        #checks random_state from 0 to 500
    from sklearn.model_selection import train_test_split
    x_train,x_test,y_train,y_test = train_test_split(ip,op,test_size=0.35,random_state=i)
    from sklearn.preprocessing import StandardScaler
    sc = StandardScaler()
    x_train = sc.fit_transform(x_train)    
    x_test = sc.fit_transform(x_test)
    from sklearn.linear_model import LogisticRegression
    lr = LogisticRegression()
    lr.fit(x_train,y_train)
    yp = lr.predict(x_test)
    print('\nFor',i,'random state:')
    from sklearn import metrics
    acc = metrics.accuracy_score(y_test,yp)
    print('Accuracy is:',acc)
    if acc>maxacc:
        maxacc=acc
        maxia=i
    pre = metrics.precision_score(y_test,yp,average='weighted')
    print('Precision is:',pre)
    if pre>maxpre:
        maxpre=pre
        maxip=i
    recall = metrics.recall_score(y_test,yp,average='weighted')
    print('Recall is:',recall)
    if recall>maxrec:
        maxrec=recall
        maxir=i
print('\n\nMax Accuracy at',maxia,'random state with Accuracy=',maxacc)
print('\nMax Precision at',maxip,'random state with Precision=',maxpre)
print('\nMax Recall at',maxir,'random state with Recall=',maxrec)
```
